```python
from langchain_community.document_loaders import PyPDFLoader

def extract_as_langchain_documents(pdf_path):
    """
    Extracts text from a PDF and returns a list of LangChain Document objects.
    """
    # Initialize the loader with the file path
    loader = PyPDFLoader(pdf_path)
    
    # Load and split the PDF into Document objects (usually one per page)
    documents = loader.load()
    
    return documents

# Example usage:
# docs = extract_as_langchain_documents("sample.pdf")
# print(f"Loaded {len(docs)} documents.")
# print(docs[0].page_content[:100])
# print(docs[0].metadata)
```

```python
docs = extract_as_langchain_documents('clinical trails.pdf')
print(docs)
```

```python
from langchain_community.document_loaders import UnstructuredPDFLoader

def load_scanned_pdf(pdf_path):
    # The 'hi_res' strategy tells it to use OCR for images
    loader = UnstructuredPDFLoader(
        pdf_path, 
        mode="elements",
        strategy="hi_res" 
    )
    
    # This returns a list of LangChain Document objects
    docs = loader.load()
    return docs

# Example usage:
# documents = load_scanned_pdf("scanned_document.pdf")
# print(documents[0].page_content)
```

```python
docs = load_scanned_pdf('clinical trails.pdf')
print(docs)
```
